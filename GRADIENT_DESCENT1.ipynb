```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red')
```

```python
#lets apply gradient descent assuming slope is constant m=78.35
# and lsts assume starting value for b is 0
y_pred=((78.35063668*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='green',label='best fit line')
plt.plot(x,y_pred,color='red',label='gradient  descent')
plt.legend()
plt.show()
```

```python
m=78.35063668
b=0
loss_slope=-2*np.sum(y-(m*x.ravel()+b))
loss_slope
```

```python
#lets take learning rate 0.1
learning_rate=0.1
stepsize=0.1*loss_slope
stepsize
```

```python
b=b-stepsize
b
```

```python
y_pred=((78.35063668*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='green',label='best fit line')
plt.plot(x,y_pred,color='red',label='gradient  descent')
plt.legend()
plt.show()
```

```python
b=100
m=78.35063668
l=0.01
epochs=10000
for i in range(epochs):
    loss_slope=-2*np.sum(y-(m*x.ravel()+b))
    stepsize=l*loss_slope
    b=b-stepsize
    y_pred=m*x+b
    plt.plot(x,y_pred,label=f'epoch {i}')
plt.scatter(x,y)
print(b)
```
